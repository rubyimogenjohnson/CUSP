```python
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point
import math
```

```python
st_x = -2.2807386
st_y = 53.4034207
point = Point(st_x, st_y) 
po = gpd.GeoSeries(point)
speed = []
direction = []
```

```python
for i in range(250):
        if i <10:
            url = 'Data/Datafile_00' + str(i) +'.csv'
        elif i<100:
            url = 'Data/Datafile_0' + str(i) +'.csv'
        else:
            url = 'Data/Datafile_' + str(i) +'.csv'
        ex_data = pd.read_csv(url)
        ex_geodata = gpd.GeoDataFrame(ex_data, geometry=gpd.points_from_xy(ex_data.x, ex_data.y))
        dis = po.geometry.apply(lambda g: ex_geodata.distance(g))
        lis = []
        for i in range(len(ex_data)):
            lis.append(dis.values[0][i])
        ex_geodata["speed"] = lis
        sp = list(ex_geodata.speed)
        speed.append(max(sp))
        x = list(ex_geodata.x)
        y = list(ex_geodata.y)
        cen_x = (min(x) + max(x))/2
        cen_y = (min(y) + max(y))/2
        tan = math.atan2((cen_y - st_y),(cen_x - st_x))
        angle = float(tan * 180 / math.pi)
        direction.append(angle)
```

```python
final = pd.DataFrame(speed)
```

```python
final["direction"] = direction
```

```python
final.columns = ['speed','direction']
```
